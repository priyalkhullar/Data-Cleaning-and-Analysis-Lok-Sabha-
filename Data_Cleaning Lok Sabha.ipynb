```python
def has_multiple_parties(row):
    parties = row['party']
    if isinstance(parties, str):  # Check if parties is a string
        party_list = parties.split(',')  # Split the string into a list of parties
        return len(party_list) >= 2
    else:
        return False
```

```python
df_multiple_parties = df[df.apply(has_multiple_parties, axis=1)]
```

```python
df_multiple_parties.head()
```

```python

def split_rows_by_party(row):

    row['party'] = str(row['party'])
    row['member'] = str(row['member'])
    row['state'] = str(row['state'])
    row['constituency'] = str(row['constituency'])
    row['constituency_type'] = str(row['constituency_type'])
    row['gender'] = str(row['gender'])

    parties = row['party'].split(',')
    members = row['member'].split(',')
    states = row['state'].split(',')
    constituencies = row['constituency'].split(',')
    constituency_types = row['constituency_type'].split(',')
    genders = row['gender'].split(',')

    num_parties = len(parties)
    members += [None] * (num_parties - len(members))
    states += [None] * (num_parties - len(states))
    constituencies += [None] * (num_parties - len(constituencies))
    constituency_types += [None] * (num_parties - len(constituency_types))
    genders += [None] * (num_parties - len(genders))

    new_rows = []
    for i in range(num_parties):
        new_row = row.copy()
        new_row['party'] = parties[i].strip()
        new_row['member'] = members[i]
        new_row['state'] = states[i]
        new_row['constituency'] = constituencies[i]
        new_row['constituency_type'] = constituency_types[i]
        new_row['gender'] = genders[i]
        new_rows.append(new_row)

    return new_rows

```

```python
# Assuming you have your DataFrame 'df1' loaded
# Apply the function to create the new DataFrame
df_split1 = pd.DataFrame([row for row_list in df1.apply(split_rows_by_party, axis=1) for row in row_list])
```
